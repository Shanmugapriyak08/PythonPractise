## Missing Values

```python
# lets capture the missing values- nan values
#1st find missing value in categorical values
feature_nan=[features for features in dataset.columns if dataset[features].isnull().sum()>1 and dataset[features].dtype=='O']
```

```python
feature_nan
```

```python
for features in feature_nan:
    print("The {} have {} % of missing values.".format(features,np.round(dataset[features].isnull().mean(),4)))
```

```python
# replace the missing value in the categorical features
def replace_cat_mis_v(dataset,feature_nan):
    data=dataset.copy()
    data[feature_nan]=data[feature_nan].fillna("Missing")
    return data

dataset=replace_cat_mis_v(dataset,feature_nan)

print(dataset[feature_nan].isnull().sum())
```

```python
# let find the missing value in the numberical features
num_fea_nan=[features for features in dataset.columns if dataset[features].isnull().sum()>1 and dataset[features].dtype!='O']
print("No .of missing value in numerical features:",len(num_fea_nan))
```

```python
for feature in num_fea_nan:
    print("The {} have {}% of missing values.".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
# we need to replace the missing value in numberical value 
# in EDA itself we found data have outliers, so we can replace nan values with either median or mode
for feature in num_fea_nan:
    median_value=dataset[feature].median()
    
    # create a new feature to capture nan values
    dataset[feature+'_nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace=True)
```

```python
def replace_num_nan(dataset,num_fea_nan):
    data=dataset.copy()
    for feature in num_fea_nan:
        median_value=data[feature].median()
        data[feature+'_nan']=np.where(data[feature],1,0)
        data[feature].fillna(median_value)

a=replace_num_nan(dataset,num_fea_nan)
```

## Numerical Variable

Numberical variable are skewed, we need to perform the log normal distribution

```python
dataset.head()
```

```python
num_feature=['LotArea','1stFlrSF','2ndFlrSF','GrLivArea','SalePrice']
for feature in num_feature:
    dataset[feature]=np.log(dataset[feature])
```

## Handling Rare Categorical Feature

We will remove categorical variables that are present less than 1% of the observations

```python
cat_fea=[feature for feature in dataset.columns if dataset[feature].dtypes=='O']
print(cat_fea)
```

```python
for feature in cat_fea:
    tmp=dataset.groupby(feature)["SalePrice"].count()/len(dataset)
    tmp_df= tmp[tmp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(tmp_df),dataset[feature],'Rare_val')
```
