```python
import pandas as pd
```

```python
df=pd.read_csv('polynomial1.csv')
df
```

```python
x=df.iloc[:,1:2].values
y=df.iloc[:,2].values
```

```python
 x
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x,y) # to load data
model.score(x,y)# to find accuracy of model
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=3)
```

```python
x_transform=poly.fit_transform(x)
```

```python
x_transform
```

```python
poly.fit(x_transform,y)
```

```python
lin=LinearRegression()
lin.fit(x_transform,y)
lin.score(x_transform,y)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y,color='blue')
plt.plot(x,model.predict(x),color='red')
plt.title('Linear Regression')
```

```python
plt.scatter(x,y,color='blue')
plt.plot(x,lin.predict(x_transform),color='green')
plt.plot(x,model.predict(x),color='red')
plt.title('Polynomial Regression')
```

```python
x=df2[['JEEPRE','JEEMAINS']]
x
```

```python
y=df2['STATUS']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model_train =model.fit(x_train,y_train)
```

```python
model_train.coef_
```

```python
model_train.score(x_train,y_train)
```

```python
pred=model_train.predict(x_test)
pred
```

```python
df3=pd.DataFrame({'Actual':y_test,
                  'Predicted':pred})
```
