```python
# Y and X variable
y=df['tip'] # targeted 
x=df[['total_bill','size']] # independent variable (input variable )
```

```python
# we need to divide the Train and Test data
```

```python
from sklearn.model_selection import train_test_split
```

```python
model=LinearRegression()
```

```python
# training the model with 80% of data
model_train=model.fit(x_train,y_train) 
```
