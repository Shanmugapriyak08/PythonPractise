# Missing values

```python
data=dataset.copy()
```

```python
feature_na=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1]
for features in feature_na:
    print("The {} have {} % of missing values.".format(features,np.round(dataset[features].isnull().mean(),4)))
```

Since they are many missing values, we need to find the relationship between missing values and Sales Price

Let's plot some diagram for this relationship

```python
for feature in feature_na:
    data=dataset.copy()
    data[feature]=np.where(data[feature].isnull(),1,0) # null is persent make entry as 1 otherwise 0
    data.groupby(feature)["SalePrice"].median().plot.bar()
    plt.title(feature)
    plt.show()
```

Here With the relation between the missing values and the dependent variable is clearly visible.So We need to replace these nan values with something meaningful which we will do in the Feature Engineering section

## Temporal Variables(Eg: Datetime Variables)

From the Dataset we have 4 year variables.
We have extract information from the datetime variables like no of years or no of days. 
One example in this specific scenario can be difference in years between the year the house was built and the year the house was sold. 
We will be performing this analysis in the Feature Engineering which is the next video.

```python
year_feature=[features for features in dataset.columns if 'Yr'in features or 'Year'in features]
year_feature
```

```python
dataset[year_feature].nunique()
```

```python
for feature in year_feature:
    print(feature,dataset[feature].unique())
    
```

```python
## Lets analyze the Temporal Datetime Variables
## We will check whether there is a relation between year the house is sold and the sales price
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel("Year sold")
plt.ylabel("Sales Price")
plt.title("Year Sold VS Sales Price")
plt.show()
```

```python
## Lets analyze the Temporal Datetime Variables
## We will check whether there is a relation between year the house is built and the sales price
dataset.groupby("YearBuilt")["SalePrice"].median().plot()
plt.xlabel("Year Built")
plt.ylabel("Sales Price")
plt.title("Year Built VS Sales Price")
plt.show()
```

```python
## Here we will compare the difference between All years feature with SalePrice
for features in year_feature:
    data=dataset.copy()
    if features!='YrSold':
        ## We will capture the difference between year variable and year the house was sold for
        data[features]=data["YrSold"]-data[features]
        plt.scatter(data[features],data["SalePrice"])
        plt.xlabel(features)
        plt.ylabel("SalePrice")
        plt.show()
```

```python
## Numerical variables are usually of 2 type
## 1. Continous variable and Discrete Variables
discrete_num_feat=[feature for feature in numberical_feature if len(dataset[feature].unique())<25 and feature not in year_feature and ["Id"]]
print("No.of. Discrete cloumn ",len(discrete_num_feat))
```

## Continuous variable 

```python
### Lets Find the realtionship between Continuous numberical and Sale PRice
continuous_num_var=[feature for feature in dataset[numberical_feature] if feature not in  discrete_num_feat+['Id']+year_feature]
print("No.of Continous variable:{}".format(len(continuous_num_var)))
```

```python
continuous_num_var
```

```python
dataset[continuous_num_var]
```

```python
for feature in continuous_num_var:
    data = dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

```python
# Since the Distribution value in continous variable are in Guassian distribution, we are converting it into logarithmic Transformation
for feature in continuous_num_var:
    data=dataset.copy()
    data[feature]=np.log(data[feature]) 
    data['SalePrice'] = np.log(data['SalePrice'])
    #plt.hist(data[feature],bins=25)
    plt.scatter(data[feature],data['SalePrice'])
    plt.show()
```
